```python
# Import
import random
from os import system, name

import pandas as pd
```

```python
# Function that draws the hangman on the screen
def display_hangman(chances):

  # Hangman stages list
  stages =  [ # stage 6 (end)
             '''
              ________
             |        |
             |        O
             |       \|/
             |        |
             |       / \\
             -
             ''',
             # stage 5
             '''
              ________
             |        |
             |        O
             |       \|/
             |        |
             |       /
             -
             ''',
             # stage 4
             '''
              ________
             |        |
             |        O
             |       \|/
             |        |
             |
             -
             ''',
             # stage 3
             '''
              ________
             |        |
             |        O
             |       \|
             |        |
             |
             -
             ''',
             # stage 2
             '''
              ________
             |        |
             |        O
             |        |
             |        |
             |
             -
             ''',
             # stage 1
             '''
              ________
             |        |
             |        O
             |
             |
             |
             -
             ''',
             # stage 0
             '''
              ________
             |        |
             |
             |
             |
             |
             -
             '''
            ]

  return stages[min(chances, 7)]

# Game function
def game():

    print("\nWelcome to th Hangman game!")
    print("Fruits, vegetables, or greens")
    print("Guess the word below:\n")

    # Words list for the game
    url = 'https://raw.githubusercontent.com/tarcilaalvess/Data_Analysis_Python/main/Hangman_Project/Data/list-en.csv'

    words = pd.read_csv(url, encoding='latin1', header=None)

    # Convert the column to list
    words = words[0].tolist()

    # Randomly chosen word
    word = random.choice(words)

    # Words letters list
    word_letters = [letter for letter in word]

    # List comprehension (game board)
    discovered_letters = ['_' for letter in word]

    # Number of chances
    chances = 6

    # Letters already guessed
    attempted_letters = []

    # Wrong guesses
    wrong_letters = []

    # Loop while number of chances is greater than zero
    while chances > 0:

        # Print hangman stage
        print(display_hangman(chances))
        print(" ".join(discovered_letters))
        print("\nRemaining chances:", chances)
        print("Wrong letters:", " ".join(wrong_letters))

        # Get input
        while True:
          guess = input("\nType a letter: ").lower()

          # Check if input is a single letter without accent
          if not guess.isalpha():
            print("Please enter a valid letter (no numbers, accents or special characters).")
            continue

          if len(guess) != 1:
            print("Enter only one letter at a time.")
            continue

          # Condition
          if guess in attempted_letters:
            print("\nYou already tried that letter. Try another!")
            continue

          break

        # Add guess to attempts
        attempted_letters.append(guess)

        # Check if guess is correct
        if guess in word_letters:
            index = 0
            print ("You guessed it!")

            # Loop
            for letter in word_letters:

              #Condicional
                if guess == letter:
                    discovered_letters[index] = letter
                index += 1

        # Decrement
        else:
            chances -= 1
            wrong_letters.append(guess)

        # If all spaces were filled, game is over
        if "_" not in discovered_letters:
            print("\nYou won, the word was:", word)
            print(display_hangman(chances))
            print(" ".join(discovered_letters))
            break

    # If chances run out
    if "_" in discovered_letters:
        chances = 0
        print("\nYou lost, the word was:", word)
        print(display_hangman(chances))
        print(" ".join(discovered_letters))
        print("Wrong letters:", " ".join(wrong_letters))


# Bloco main
if __name__ == "__main__":
    game()
```
